```python
import matplotlib.pyplot as plt

day = [1,2,3,4,5,6,7]
hours = [4,6,3,9,10,1,6]

colors = ["r","g","b","c","m","y","k"]
sizes = [400,520,380,470,695,502,510]
mark = ["*","#","$","%","^","@","!"]

plt.scatter(day,hours,c=colors,s=sizes,alpha=0.9,marker="*",edgecolor="g",linewidth=2)
plt.title("Scatter Plot",fontsize = 15)
plt.xlabel("Day No.",fontsize=10)
plt.ylabel("Hours Worked",fontsize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1,2,3,4,5,6,7]
hours = [4,6,3,9,10,1,6]
wages = [10,23,14,25,12,23,18]

colors = ["r","g","b","c","m","y","k"]
sizes = [400,520,380,470,695,502,510]
mark = ["*","#","$","%","^","@","!"]

plt.scatter(day,hours,c=colors,s=sizes,alpha=0.9,marker="*",edgecolor="g",linewidth=2)
plt.scatter(day,wages,c=colors,s=sizes,alpha=0.9,marker=".",edgecolor="r",linewidth=2)
plt.title("Scatter Plot",fontsize = 15)
plt.xlabel("Day No.",fontsize=10)
plt.ylabel("Hours Worked",fontsize=10)
plt.show()
```
