```python
import numpy as np
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [10,20,30,40,50,60]
z = [5,10,15,20,25,30]

width = 0.2

p = np.arange(len(x))
q = [j+width for j in p]


plt.bar(p,y,width,color="r",label="10 times")
plt.bar(q,z,width,color="g",label="5 times")

plt.xticks(p+width/2 , x , rotation=15)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [10,20,30,40,50,60]
z = [5,10,15,20,25,30]

width = 0.2

p = np.arange(len(x))
q = [j+width for j in p]


plt.barh(p,y,width,color="r",label="10 times")
plt.barh(q,z,width,color="g",label="5 times")

plt.xticks(p+width , x , rotation=15)
plt.legend()
plt.show()
```
