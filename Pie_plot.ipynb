```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ["C","C++","Java","Python"]

plt.pie(x,labels=y,labeldistance=0.8)
plt.show()
```

```python
x = [10,20,30,40]
y = ["C","C++","Java","Python"]
ex = [0.0,0.0,0.5,0.0]
c = ["r","g","b","m"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,radius=1.2,labeldistance=1.1,startangle=90)
plt.show()
```

```python
x = [10,20,30,40]
y = ["C","C++","Java","Python"]
ex = [0.0,0.0,0.3,0.0]
c = ["r","g","b","m"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",textprops={"fontsize":15},counterclock=False,rotatelabels=True)
plt.title("Languages")
plt.legend(loc=2)
plt.show()
```
