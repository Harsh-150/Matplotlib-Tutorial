```python
import matplotlib.pyplot as plt

x = ["Java","Javascript","C++","Python","Node.js"]
y = [90,93,78,94,86]

plt.xlabel("Language" , fontsize=15)
plt.ylabel("Popularity" , fontsize=15)

plt.title("Popularity of Programming Languages" , fontsize=20)

colors = ["y","b","r","g","k"]

plt.bar(x,y,width=0.6,color=colors,align="edge",edgecolor="m",linewidth=4,linestyle=":",alpha=0.6) # align = "center" is also there
plt.show()
```

```python
Professions = ["Lawyer","Doctor","Engineer","Pilot","Civil Servant"]
Popularity = [65,82,92,45,99]

plt.xlabel("Profession")
plt.ylabel("Popularity")

plt.title("Popularity of Careers")

colors = ["y","b","r","g","c"]

plt.bar(Professions,Popularity,color=colors,label="Career Opportunities")
plt.legend()
plt.show()
```
